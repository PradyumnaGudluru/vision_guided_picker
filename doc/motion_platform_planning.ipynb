```python
import numpy as np
from scipy.linalg import logm, expm
```

```python
import matplotlib.pyplot as plt


dt = 1 / 60

v = np.random.randn(4096)
x = np.cumsum(v) * dt

fig, ax = plt.subplots()

ax.plot(x)
```

```python
acc = np.random.randn(8192)

y = (0, 0)
gamma = 1e-3
k = 1e-3

xs = []

for a in acc:
    x, v = y
    y = (x + v * dt, a * dt - gamma * v - k * x)
    xs.append(x)
    
fig, ax = plt.subplots()
ax.plot(xs)
ax.plot(np.cumsum(np.cumsum(acc)) * dt**2)
```
